# Global Internet Users Dataset
The following will be an exploration of the global internet users dataset. I will create a few illustrative plots that will highlight some aspects of the data


Source: https://www.kaggle.com/datasets/ashishraut64/internet-users

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the data
data = pd.read_csv('data.csv', index_col='Year')
```

```python
data.head()
```

Create Violin plot:

```python
# Make a violin plot where the x-axis is the Year and the y-axis is the %of internet users from 0 to 100
# get the input data by getting the column 'Internet Users (%)' as a numpy array for each year
year_list = sorted(np.unique(data.index))


input_data = [data.loc[year, 'Internet Users(%)'].values for year in year_list]
# Create the violin plot
fig, ax = plt.subplots()
ax.violinplot(input_data, showmedians=True)
```

Make the plot bigger and add x-labels

```python
# Make a violin plot where the x-axis is the Year and the y-axis is the %of internet users from 0 to 100
# get the input data by getting the column 'Internet Users (%)' as a numpy array for each year
year_list = sorted(np.unique(data.index))


input_data = [data.loc[year, 'Internet Users(%)'].values for year in year_list]
# Create the violin plot
fig, ax = plt.subplots(figsize=(18, 8))
ax.violinplot(input_data, showmedians=True)


# set the x-labels to be yearlist
ax.set_xticks(np.arange(1, len(year_list) + 1))
ax.set_xticklabels(year_list, rotation=45)
k=0
```

Trim away every 2nd year to add more meaningful detail

```python
# Make a violin plot where the x-axis is the Year and the y-axis is the %of internet users from 0 to 100
# get the input data by getting the column 'Internet Users (%)' as a numpy array for each year
year_list = sorted(np.unique(data.index))

# make year_list to be every second year
year_list = year_list[::2]

input_data = [data.loc[year, 'Internet Users(%)'].values for year in year_list]
# Create the violin plot
fig, ax = plt.subplots(figsize=(18, 8))
ax.violinplot(input_data, showmedians=True)

# set the x-labels to be yearlist
ax.set_xticks(np.arange(1, len(year_list) + 1))
ax.set_xticklabels(year_list, rotation=45)

k=0
```

Add Labels and gid lines

```python
# Make a violin plot where the x-axis is the Year and the y-axis is the %of internet users from 0 to 100
# get the input data by getting the column 'Internet Users (%)' as a numpy array for each year
year_list = sorted(np.unique(data.index))

# make year_list to be every second year
year_list = year_list[::2]

input_data = [data.loc[year, 'Internet Users(%)'].values for year in year_list]
# Create the violin plot
fig, ax = plt.subplots(figsize=(18, 8))
ax.violinplot(input_data, showmedians=True)

# format the plot
ax.set_xticks(np.arange(1, len(year_list) + 1))
ax.set_xticklabels(year_list, rotation=45)
ax.set_xlabel('Year', fontsize=16)
ax.set_ylabel('Internet Users (%)', fontsize=16)
ax.set_title('Distribution of Internet Users (%) by Year', fontsize=16)

# add gridlines and make them light color
ax.grid(axis='y', color='lightgray')
```

Highlight the Median and add a small description box

```python
# Make a violin plot where the x-axis is the Year and the y-axis is the %of internet users from 0 to 100
# get the input data by getting the column 'Internet Users (%)' as a numpy array for each year
year_list = sorted(np.unique(data.index))

# make year_list to be every second year
year_list = year_list[::2]

input_data = [data.loc[year, 'Internet Users(%)'].values for year in year_list]
# Create the violin plot
fig, ax = plt.subplots(figsize=(24, 8))
violin_plot_reference = ax.violinplot(input_data, showmedians=True)

# format the plot
ax.set_xticks(np.arange(1, len(year_list) + 1))
ax.set_xticklabels(year_list, rotation=45)
ax.set_xlabel('Year', fontsize=16)
ax.set_ylabel('Internet Users (%)', fontsize=16)
ax.set_title('Distribution of Internet User Percentage across Countries', fontsize=16)

# add gridlines and make them light color
ax.grid(axis='y', color='lightgray')

violin_plot_reference['cmedians'].set_color('red')
violin_plot_reference['cmedians'].set_linewidth(5)

# Add Legend indicating that the median is the red line
ax.legend([violin_plot_reference['cmedians']], ['Median Internet-Using Population Percentage'], loc='upper left', fontsize=16)

# Add a text box describing the bimodal distribution
text = "In recent years, the distribution appears to \nbecome bimodal,with countries having a\nhigh or low percentage of internet users."
ax.text(0.02, 0.3, text, transform=ax.transAxes, fontsize=12, verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
```
